# Exercise 02 — Relacionamento entre variáveis

O objetivo desse exercício é desenvolver a capacidade de relacionar variáveis e formular interpretações iniciais, utilizando comparações simples, filtros e observações visuais. Sem métricasa vançadas ou análises estatísticas formais.

## Carregamento e revisão do dataset

```python
import pandas as pd

df = pd.read_csv("../datasets/raw/tips.csv")

df.head(5)
```

```python
df.info()
```

Quantitativas: `total_bill, tip, size` | Categóricas: `sex, smoker, day, time`.

## Relação entre variáveis quantitativas

```python
import matplotlib.pyplot as plt

plt.scatter(df["total_bill"], df["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

Observa-se uma relação positiva entre total_bill e tip, indicando que contas mais altas tendem a estar associadas a gorjetas maiores. A dispersão sugere que, embora exista uma tendência clara, a gorjeta não é uma proporção fixa do valor total, havendo variabilidade entre as observações.

```python
plt.scatter(df["total_bill"], df["size"])
plt.xlabel("Total Bill")
plt.ylabel("Party Size")
plt.title("Total Bill vs Party Size")
plt.show()
```

Observa-se que não há uma relação clara entre size e total_bill. Os pontos estão bastante dispersos, indicando que o aumento no número de pessoas na mesa não implica necessariamente em contas mais altas.

## Variáveis quantitativas x categoricas

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("Distribuição do total da conta por dia")
plt.show()
```

Observa-se que os valores medianos do total da conta são semelhantes entre os dias da semana, indicando que, de forma geral, o gasto típico não varia significativamente por dia. No entanto, a presença de outliers em alguns dias indica a ocorrência ocasional de contas muito altas, sugerindo uma assimetria à direita na distribuição.

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x="time", y="total_bill")
plt.title("Distribuição do total da conta por período")
plt.show()
```

Observa-se que os valores medianos do total da conta são semelhantes entre almoço e jantar, sugerindo que o período do dia não impacta significativamente o valor típico da conta. Ambos os períodos apresentam outliers associados a contas mais elevadas, indicando assimetria à direita nas distribuições.

## Observações guiadas

A relação entre as variáveis total_bill e tip apresenta uma tendência positiva visualmente consistente, sugerindo que valores mais altos de conta tendem a estar associados a gorjetas maiores. Embora essa observação seja compatível com a hipótese de que a gorjeta possa representar uma proporção do valor total da conta, não há evidência suficiente neste momento para confirmar essa relação, sendo esta apenas uma hipótese baseada em observação visual e intuição externa ao dataset.

Por outro lado, a relação entre o tamanho da mesa (size) e o valor total da conta (total_bill) não se mostrou visualmente clara. Os pontos encontram-se dispersos, sem um padrão evidente que indique que o aumento no número de clientes esteja associado, de forma consistente, a um aumento no valor total da conta. Isso sugere que, neste conjunto de dados, essa relação não pode ser inferida apenas por observação gráfica.

Ao analisar a distribuição do valor total da conta por dia da semana, observa-se que as medianas são relativamente semelhantes entre os diferentes dias, indicando pouca variação central. Apesar da presença de alguns valores atípicos que podem influenciar a média, a mediana sugere que o dia da semana, por si só, não parece impactar fortemente o valor da conta, ao menos sob uma análise exploratória inicial.

De forma semelhante, a distribuição do valor total da conta por período (lunch e dinner) mostra medianas bastante próximas. Embora o período do jantar concentre um maior número de observações, isso não se traduz, visualmente, em diferenças significativas no valor típico das contas.
