# Exercise 03 — Distribuições e padrões iniciais

Neste exercício, o objetivo é analisar a distribuição das variáveis do dataset **Tips**, buscando identificar padrões iniciais, assimetrias e valores extremos.

O foco está na **observação visual das distribuições**, sem o uso de métricas agregadas ou inferências estatísticas.  
A intenção é desenvolver a capacidade de interpretar dados a partir de sua forma e dispersão.

```python
import pandas as pd

df = pd.read_csv("../datasets/raw/tips.csv")

df.head(5)
```

## Distribuição de variáveis quantitativas

```python
import matplotlib.pyplot as plt

plt.hist(df["total_bill"], bins=20)
plt.xlabel("Total da conta")
plt.ylabel("Frequência")
plt.title("Distribuição do total da conta")
plt.show()
```

Observa-se que a distribuição do valor total da conta apresenta maior concentração de valores entre aproximadamente 5 e 30. Há a presença de poucos valores mais elevados, que se estendem para a direita da distribuição, indicando uma cauda à direita. Esse comportamento sugere uma distribuição assimétrica positiva.

```python
plt.hist(df["tip"], bins=20)
plt.xlabel("Gorjeta")
plt.ylabel("Frequência")
plt.title("Distribuição da gorjeta")
plt.show()
```

A distribuição da gorjeta mostra uma concentração de valores principalmente entre 1 e 4. Observa-se também a presença de alguns valores mais altos, menos frequentes, que se estendem à direita da distribuição. Esse padrão indica uma assimetria à direita, semelhante à observada na variável total da conta.

```python
plt.hist(df["size"], bins=20)
plt.xlabel("Número de pessoas na mesa")
plt.ylabel("Frequência")
plt.title("Distribuição do tamanho da mesa")
plt.show()
```

A distribuição do tamanho da mesa apresenta forte concentração no valor 2, indicando que a maioria das mesas era composta por duas pessoas. À medida que o número de pessoas aumenta, a frequência diminui rapidamente, formando uma cauda à direita mais pronunciada. Por se tratar de uma variável discreta e com poucos valores possíveis, a assimetria é mais evidente em comparação às variáveis contínuas analisadas anteriormente.

## Comparação de distribuições entre grupos

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("Distribuição do total da conta por dia")
plt.show()
```

Observa-se que, para a maioria dos dias da semana, a distribuição do valor total da conta apresenta assimetria à direita, caracterizada pela presença de valores extremos elevados. A exceção aparente é a sexta-feira (Fri), cuja distribuição parece mais próxima da simetria.

As caixas são relativamente compactas na maioria dos dias, indicando baixa dispersão dos valores centrais (entre o primeiro e o terceiro quartil). As medianas dos dias Sun e Sat são visualmente semelhantes, enquanto Thur e Fri apresentam medianas ligeiramente distintas, embora sem grandes deslocamentos.

Todos os grupos exibem outliers à direita, sugerindo a ocorrência de poucas contas com valores significativamente mais altos, o que contribui para a assimetria positiva observada nas distribuições.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="time", y="total_bill")
plt.title("Distribuição do total da conta por período do dia")
plt.show()
```

Tanto no período de Lunch quanto de Dinner, observa-se uma distribuição assimétrica à direita, com presença de valores extremos elevados. As caixas são relativamente compactas em ambos os períodos, indicando concentração dos valores centrais.

As medianas apresentam uma leve diferença entre os períodos, embora não seja possível afirmar uma diferença relevante apenas com base na observação visual. Apesar de parecer haver mais observações no grupo Dinner, o boxplot não permite conclusões sobre frequência, apenas sobre a forma e a dispersão da distribuição.

## Assimetria e valores extremos

Observa-se uma forte concentração dos valores de total_bill em faixas intermediárias, enquanto poucos valores elevados se estendem para a extremidade direita da distribuição. Esses valores extremos, embora raros, criam uma cauda longa, indicando assimetria positiva. A presença desses extremos sugere que medidas de tendência central podem ser influenciadas por poucos registros fora do padrão observado.

Na distribuição da gorjeta, nota-se uma maior concentração em faixas iniciais e intermediárias, enquanto valores mais elevados se estendem para a extremidade direita da distribuição. Esses valores, apesar de pouco frequentes, criam uma cauda longa, indicando assimetria positiva. Assim como observado em total_bill, esses valores extremos podem influenciar a interpretação de medidas de tendência central.

A distribuição do tamanho da mesa apresenta forte concentração no valor 2, indicando que a maioria das mesas possui dois clientes. Alguns poucos valores mais elevados se estendem para a extremidade direita da distribuição, formando uma cauda longa. A assimetria positiva nessa variável é mais perceptível, reforçando o caráter discreto e concentrado da distribuição.

## Observações guiadas

Observa-se que a variável size apresenta maior assimetria em sua distribuição quando comparada a total_bill e tip, devido à forte concentração em poucos valores discretos e à presença de uma cauda longa à direita. A partir das distribuições observadas, não é possível tirar conclusões adicionais sem aprofundar a análise.

Nenhuma das comparações entre grupos pode ser considerada enganosa neste estágio, uma vez que não há base suficiente para inferir padrões causais ou tomar decisões a partir dos dados observados. O boxplot mostrou-se uma ferramenta eficaz para identificar valores extremos de forma clara, enquanto o histograma, apesar de útil para visualizar a distribuição geral, pode ocultar outliers quando estes são pouco frequentes.
