# Preparação do Dataset — Tips

Este notebook é responsável pela geração e preparação inicial do dataset utilizado
nos exercícios da Week 03.

⚠️ Este notebook deve ser executado apenas uma vez.
Os notebooks de exercícios devem carregar os dados exclusivamente a partir do
diretório `datasets/processed/`.

```python
from pathlib import Path

raw_path = Path("datasets/raw/tips.csv")
processed_path = Path("datasets/processed/tips_clean.csv")

if raw_path.exists() or processed_path.exists():
    raise FileExistsError(
        "O dataset já foi gerado. Este notebook não deve ser reexecutado."
    )
```

## Geração do dataset raw

```python
import seaborn as sns

tips = sns.load_dataset("tips")

tips.to_csv(raw_path, index=False)
```

## Carregamento e limpeza básica

```python
import pandas as pd

df = pd.read_csv(raw_path)

# Verificações básicas
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
# Remoção de duplicatas, se houver
df = df.drop_duplicates()
```

## Salvamento do dataset processed

```python
df.to_csv(processed_path, index=False)
```
